```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import json
import requests
import time
from scipy.stats import linregress
from scipy import stats

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
units = "imperial"
api_key = weather_api_key
#query_url = url + "appid=" + api_key + "&q=" + city
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key
#query_url = f"{url}appid={api_key}&units={units}&q="
#passing url any string (the &q= the city passes through) url does not change but the city does
#I have to pass the city through a for loop
response = requests.get(f"{url}&q={city}").json() 
response
```

having a list

```python
# Create empty lists to append the API data into lists 
city_name = []
cloudiness = []
country = []
date = []
humidity = []
latitude = []
longitude = []
max_temp = []
wind_speed = []

# Start the call counter 
record = 0
# Log file print statement
print(f"Beginning Data Retrieval")
print(f"-------------------------------")
#Loop through the cities in the city list
#things can fail quietly
for city in cities:
    url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key
    try:
        response = requests.get(f"{url}&q={city}").json()
        city_name.append(response['name'])
         #'clouds': {'all': 100}
        cloudiness.append(response['clouds']['all'])
        #'sys': {'country': 'ID', 'sunrise': 1587597989, 'sunset': 1587642043}
        country.append(response['sys']['country'])
        ##'dt': 1587588902,
        date.append(response['dt'])
        #'humidity': 89,
        humidity.append(response['main']['humidity'])
        #{'coord': {'lon': 96.13, 'lat': 4.14},
        latitude.append(response['coord']['lat'])
        longitude.append(response['coord']['lon'])
         #'temp_max': 77.7,
        max_temp.append(response['main']['temp_max'])
        #'wind': {'speed': 4.05, 'deg': 86},
        wind_speed.append(response['wind']['speed'])
        city_record = (response['name'])
        record += 1
        print(f"Processing Record Response {record} : {city_record}")
        #print(f"{url}&q={city}")
        time.sleep(1)
        
        
    except:
     print("City not Found. Skipping")
    
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#Make a data frame from the calls
city_dict = {
    "City": cities,
    "Cloudiness": cloudiness,
    "Country": country,
    "Date": date,
    "Humidity": humidity,
    "lat" : latitude,
    "Lng": longitude,
    "max_temp": max_temp,
    "Wind Speed": wind_speed}
#the below code line did not work,so I had to do tranpose to get the code to work
#city_data = pd.DataFrame(city_dict)
city_data = pd.DataFrame.from_dict(city_dict, orient='index')
city_data.transpose()
city_data.head()
#switch the columns and rows
city_data = city_data.T
city_data
city_data.to_csv('city_data.csv')
city_data.head()
```

## Linear Regression

```python
# Create Northern and Southern Hemisphere DataFrames
#northern_hem = city_data.loc[city_data["lat"] >= 0]
#northern_hem.reset_index(inplace=True)


northern_df = city_data.query("lat >= 0")
northern_df = northern_df.reset_index(drop=True)

southern_df = city_data.query("lat < 0")
southern_df = southern_df.reset_index(drop=True)
#southern_hem = city_data.loc[city_data["lat"] < 0]
#southern_hem.reset_index(inplace=True)
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#creating linear regression for northern hem vs. max temp
x_axis = northern_df["lat"].astype(float)
y_axis = northern_df["max_temp"].astype(float)


#regression lines
#weather_stats activity

(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(x_axis, y_axis)
regress_values = x_axis * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))



plt.scatter(x_axis, y_axis, marker="o")
plt.title("Northern Hemisphere- Latitude vs. Max Temperature")
plt.annotate(line_eq,(20,15),fontsize=15,color="red")
plt.plot(x_axis, regress_values,color="red")
plt.ylabel("Max Temperature")
plt.xlabel("Latitude")
plt.ylim(0,100)
plt.xlim(0,80)
plt.grid(True)
# Save the figure
plt.savefig("northernhemlat_vs_max_temp.png")

# Show plot
plt.show()
```
